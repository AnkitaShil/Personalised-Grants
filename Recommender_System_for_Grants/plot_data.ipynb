```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
df_in = pd.read_csv("kiva_loans.csv//recep_data.csv")
df_in.head()
lender = pd.read_csv("kiva_loans.csv//lender_data.csv")
lender.head()
```

```python
df_in.describe()
```

```python
pca_2d = PCA(n_components = 2)
df_pca_2d = pca_2d.fit_transform(df_in)
lender_pca_2d = pca_2d.fit_transform(lender)
```

```python
pca_3d = PCA(n_components = 3)
df_pca_3d = pca_3d.fit_transform(df_in)
lender_pca_3d = pca_3d.fit_transform(lender)
```

```python
fig3d = plt.figure(1)
ax = fig3d.add_subplot(111, projection='3d')
ax.scatter(df_pca_3d[:, 0], df_pca_3d[:, 1], df_pca_3d[:, 2])
ax.scatter(lender_pca_3d[:, 0], lender_pca_3d[:, 1], lender_pca_3d[:, 2])
```

```python
plt.scatter(df_pca_2d[:, 0], df_pca_2d[:, 1])
plt.scatter(lender_pca_2d[:, 0], lender_pca_2d[:, 1])
```
